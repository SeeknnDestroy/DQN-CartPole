````python
from pathlib import Path

from dqn_cartpole.config import DQNConfig
from dqn_cartpole.evaluate import evaluate_checkpoint
from dqn_cartpole.train import train
````

````python
artifacts_dir = Path("artifacts/notebook_demo")
config = DQNConfig(
    episodes=25,
    batch_size=32,
    replay_buffer_size=5000,
    hidden_sizes=(64, 64),
    checkpoint_path=artifacts_dir / "cartpole_dqn.pt",
    metrics_path=artifacts_dir / "train_metrics.json",
    log_every=5,
)

training_metrics = train(config)
training_metrics["best_moving_average_reward"]
````

````python
evaluation_metrics = evaluate_checkpoint(
    checkpoint_path=config.checkpoint_path,
    episodes=10,
    seed=config.seed + 100,
    metrics_path=artifacts_dir / "eval_metrics.json",
)
evaluation_metrics
````
